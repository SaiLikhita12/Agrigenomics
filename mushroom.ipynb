```python
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
d = defaultdict(LabelEncoder)
Xfit = X.apply(lambda x: d[x.name].fit_transform(x))
```

```python
Xfit.shape
```

```python
Xfit
```

```python
le_y = LabelEncoder()
yfit = le_y.fit_transform(y)
```

```python
yfit
```

```python
from sklearn.model_selection import train_test_split
X_inp_train,X_test,y_inp_train,y_test = train_test_split(Xfit,yfit,test_size=0.1,random_state=1)
```

```python
X_inp_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_inp_train,y_inp_train)
y_pred = logreg.predict(X_test)
```

```python
y_pred = logreg.predict(X_test)
y_pred
```

```python
logreg.score(X_inp_train,y_inp_train)
```

```python
from sklearn import metrics
print("Accuracy :",(metrics.accuracy_score(y_test,y_pred))*100)
print("Precision:",metrics.precision_score(y_test,y_pred))
print("Recall: ",metrics.recall_score(y_test,y_pred))
```
